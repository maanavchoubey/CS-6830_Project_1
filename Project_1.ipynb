According to Wikipedia, a pitcher commits a balk when he pretends to pitch but does not actually toss the ball. According to the graph, a pitcher would commit fewer than one balk every season on average. But what in the world happened in 1988, when the average number of balks per pitcher rose to almost 2.0? For better comprehension, let's examine the year 1988 in further depth.

```python
pitcher_df[pitcher_df.yearID == 1988].BK.hist(bins=np.arange(0,16,2), alpha=0.5, figsize=(10,5))
pitcher_df[pitcher_df.yearID == 1987].BK.hist(bins=np.arange(0,16,2), alpha=0.5, figsize=(10,5))
plt.xlim(0,16)
plt.title("A comparison in average number of balks in 1987(orange) and 1988(blue)", size=15)
plt.ylabel('number of pitchers')
plt.xlabel('number of balks')
plt.show()
```

In the year 1988, something is going on. When I typed "major league balks 1988" into Google, I got a lot of results. Google came up with these results: http://reconditebaseball.blogspot.kr/2008/08/balks-story-of-1988-major-league.html The Story of the 1988 Major League Baseball Season! is the title of the blog post. This sounds really interesting!

The only difference between the two rules is that "complete stop" was changed to "single complete and clear stop, with both feet on the ground" in the 1988 version. The goal of this small change was to make balk calls more consistent across all of Major League Baseball. However, it led to one of the most frustrating summers ever for pitchers in Major League Baseball. Rick Mahler of the Atlanta Braves balked 357 times in 1988, which was only six weeks after the season started. He broke the MLB record for most balks in a full season, even though there were still three-quarters of the season to go.

```python
batter_df = pd.read_csv(dir+"Batting.csv")
```

```python
batter_df.head()
```

```python
players_df = pitcher_df[['playerID','yearID']].append(batter_df[['playerID','yearID']])
```

```python
people_df = pd.read_csv(dir+'People.csv')
```

```python
people_df.columns
```

```python
people_df.head()
```

```python
players_nat_df = players_df.merge(people_df, on="playerID", how='left')[['playerID','yearID','birthCountry']]
```

```python
players_nat_df.head()
```

```python
nat_result = players_nat_df.groupby(['yearID','birthCountry'], as_index=False).playerID.count()
```

```python
nat_result.head()
```

```python
def ratio_calc(x):
    total = x.sum()
    return x / total
```

```python
nat_result['ratio'] = nat_result.groupby('yearID')['playerID'].apply(ratio_calc)
```

```python
nat_result.pivot_table(index='yearID', columns='birthCountry', values='ratio').plot(figsize=(20,6))
plt.legend().set_visible(False)
plt.title('MLB Nationality Diversity over Time in percentage')
plt.show()
```

Since 1940, the number of US players has been going down, and by the end of the 1990s, only 80% of players were from the US. As of 2015, about 76% of the players are from the United States.

```python
#Excluding USA to have a look at other countries.
nat_result_wo_usa = nat_result[nat_result.birthCountry != 'USA']
nat_result_wo_usa.pivot_table(index='yearID', columns='birthCountry', values='ratio').plot(figsize=(20,6))
plt.legend().set_visible(False)
```

```python
# Picking up the top 10 countries in 2020, and see their records.
top10 = nat_result_wo_usa[nat_result_wo_usa.yearID == 2020].groupby('birthCountry').playerID.sum().sort_values(ascending=False).nlargest(10)
top10
```

```python
top10_list = top10.index.tolist()
```

```python
def in_top10(nat):
    if nat in top10_list:
        return True
    else:
        return False
```

```python
nat_result['top10'] = nat_result.birthCountry.apply(lambda x: in_top10(x))
nat_result.head()
```

```python
nat_result[nat_result.top10].pivot_table(index='yearID', columns='birthCountry', values='ratio').plot(figsize=(20,6))
```

Still another cool chart! From 1950 to 1970, there were a lot of Cuban players, but since 1980, the number of Cuban players has been going down. This is probably because of tense diplomatic issues between two countries, or the players may have moved to the United States. And there are a lot of D.R. players in the MLB (Dominican Republic and Venezuela). Puerto Rico has an interesting trend: its percentage grew quickly in the 1990s (it was in second place), but it has been slowly going down since 2000 and is now about the same as Japan and Mexico.

```python
teams_df = pd.read_csv(dir+"Teams.csv")
```

```python
#Review if a World Series was played in any particular year
years_ws = teams_df[['yearID','WSWin']].groupby('yearID', as_index=False).sum()
```

```python
#Create a list of years where a World Series was played and won

years_ws = years_ws[years_ws['WSWin'].str.contains('Y', na=False)]
years_ws.head()
```

```python
teams_df.drop(['lgID','Ghome','divID','DivWin','WCWin','LgWin','IPouts','HA','HRA','BBA','SOA','E','DP','FP','park',
                     'attendance','BPF','PPF','CS','HBP','SF', 'teamIDBR','teamIDlahman45','teamIDretro'], 
                    axis=1,
                    inplace=True)
```

```python
teams_df = teams_df[teams_df['yearID'].isin(years_ws['yearID'])]

teams_df.head()
```

The data on salaries starts in 1985. The teams df should show teams from 1985 on (years prior should be removed)

```python
teams_df = teams_df[teams_df['yearID'] >= 1985]
teams_df['yearID'].describe()
```

```python
# Review total number of runs scored each year
mean_runs_df = teams_df.groupby('yearID', as_index=False)['R'].mean()
mean_runs_df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(mean_runs_df['yearID'],mean_runs_df['R'], color = 'b')
plt.title('Mean Runs Scored by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Runs', fontsize=12);
```

```python
# Combine the Salary and Runs graphs
fig, ax1 = plt.subplots(figsize=(10,6))

ax1.set_xlabel('Year')
ax1.set_ylabel('Mean Salary')
ax1.plot(mean_salary_df.groupby('yearID').sum(), color='r')

ax2 = ax1.twinx()

ax2.set_ylabel('Runs')
ax2.plot(mean_runs_df['yearID'], mean_runs_df['R'], color = 'b');
```

Run production peaked in the late 1990s and early 2000s, when Mark McGuire, Sammy Sosa, and Barry Bonds were at the top of their games. When the MLB started testing for performance-enhancing drugs (PEDs) more closely in the early 2000s, there was a noticeable drop in the number of runs scored each year. But notice that even though the number of runs scored each year went down, salaries kept going up.

```python
salaries_2016 = salaries[salaries['yearID'] == 2016]
merge_all = pd.merge(people_df,salaries_2016,on = ['playerID'])
finalData = merge_all.drop(["birthDay","birthCountry","birthState","birthCity","deathYear","deathMonth","deathDay","deathCountry","deathState","deathCity","nameFirst","nameLast","nameGiven","debut","finalGame","bbrefID","yearID","lgID","retroID"],axis = 1)
print(len(finalData))
print(len(people_df))
print(len(salaries))
```

```python
salaries_birthYear = finalData.groupby('birthYear').mean()
salaries_birthYear.plot(y= 'salary')
plt.ylabel('salary')
plt.title('Salaries and BirthYear')
salaries_birthMonth = finalData.groupby('birthMonth').mean()
salaries_birthMonth.plot(y = 'salary')
plt.ylabel('salary')
plt.title('Salaries and BirthMonth')
```

The players who make the most money are around the year 1975. The average age of these players is about 43. The least money goes to players born in June. They could be players who join the team in June. The fact that they are in the initial stage of their carrer will also effect their future development.

```python
salaries_weight = finalData.groupby('weight').mean()
salaries_weight.plot(y = 'salary')
plt.ylabel('salary')
plt.title('Salaries and weight')

salaries_height = finalData.groupby('height').mean()
salaries_height.plot(y = 'salary')
plt.ylabel('salary')
plt.title('Salaries and height')
```

The salary goes up the more height and weight there are. It's more likely that people will make a lot of money.
